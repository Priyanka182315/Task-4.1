```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "twitter_training.csv"
df = pd.read_csv(file_path)

# Rename columns for clarity
df.columns = ["ID", "Topic", "Sentiment", "Text"]

# Set visualization style
sns.set_style("whitegrid")

# Count sentiment distribution
sentiment_counts = df["Sentiment"].value_counts()

# Plot sentiment distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=sentiment_counts.index,
            y=sentiment_counts.values, 
            hue=sentiment_counts.index,  # Assign x variable to hue 
            palette="coolwarm",
            legend=False) # Disable legend to maintain the same effect 
plt.title("Sentiment Distribution", fontsize=14)
plt.xlabel("Sentiment", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45)
plt.show()

# Count sentiment per topic
topic_sentiment_counts = df.groupby(["Topic", "Sentiment"]).size().unstack().fillna(0)

# Plot sentiment distribution by topic
plt.figure(figsize=(12, 6))
topic_sentiment_counts.plot(kind="bar", stacked=True, colormap="coolwarm", figsize=(12, 6))
plt.title("Sentiment Distribution by Topic", fontsize=14)
plt.xlabel("Topic", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Sentiment")
plt.show()


```
